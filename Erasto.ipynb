```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
dump_site = pd.read_csv("https://data.wprdc.org/dataset/ed4cf849-d323-4d01-8a13-675d283f1ac9/resource/ee834d8d-ae71-4b3b-b02b-312ba321ff17/download/illegaldumpsites.csv")
dump = dump_site.loc[dump_site['City'] == 'Pittsburgh']
dumps = dump.drop(['site_name', 'Status','City','location_description','latitude','longitude'], axis = 1)
new_data = dumps.iloc[:,0:2]
new_data['estimated_tons']=new_data['estimated_tons'].astype(float)
new = new_data.groupby('Neighborhood', as_index = False).sum()
new.head(79)
```

Removed all unnecessary datasets that didn't have anything to do with our data. There is 78 different neighborhoods in our dataset so we included them all.

```python
new.plot(x='Neighborhood', kind = 'bar', rot=90, figsize =(30,20))
```

The 5 worst neighborhoods in terms of trash are Beltzhover, Carrick, Duck Hollow, Hazelwood and Perry South.

```python
best = new.loc[new['estimated_tons'] < .8]
best.plot(x='Neighborhood', kind = 'bar', rot=90, figsize =(30,20))
```

East Libery and Springdale are the cleanest with 0 tons, followed by Point Breeze North and Strip District with approx. .3 tons, followed by Homewood West with 0.5 tons.
